```python
# Exemplo 1

input = """............
........0...
.....0......
.......0....
....0.......
......A.....
............
............
........A...
.........A..
............
............"""

# with open("C:/Projetos/Advent Of Code/2024/puzzle inputs/7.txt", "r") as arquivo:
#     input = arquivo.read()

matrix = [list(row) for row in input.split("\n")]
antennas = []

for i, linha in enumerate(matrix):
    for j, caractere in enumerate(linha):
        if caractere not in [".", "#"]:
            antennas.append([caractere, (i, j)])

def find_antinodes_location(antennas):
    antinodes = []
    for i, antenna in enumerate(antennas):
        for j, other_antenna in enumerate(antennas):
            if antenna[0] == other_antenna[0] and i != j:
                antinode_X, antinode_Y = (antenna[1][0] - other_antenna[1][0]), (antenna[1][1] - other_antenna[1][1])
                newAntinode = (antenna[0], (antenna[1][0] + antinode_X, antenna[1][1] + antinode_Y))
                antinodes.append(newAntinode)

    return list(set(antinodes))

def paint_antinodes(antinodes):
    count_antinodes = 0

    for antinode in antinodes:
        x, y = antinode[1]

        if x >= 0 and x < len(matrix) and y >= 0 and y < len(matrix[0]):
            if matrix[x][y] == "." or matrix[x][y] == '#':
                matrix[x][y] = "#"
                count_antinodes += 1

    return count_antinodes

print(paint_antinodes(find_antinodes_location(antennas)))
```

```python
# CÃ³digo 1

rows = []

with open("C:/Projetos/Advent Of Code/2024/puzzle inputs/8.txt", "r") as file:
    content = file.readlines()
    for line in content:
        rows.append(line.strip())

positions = {}

for x, row in enumerate(rows):
    for y, c in enumerate(row):
        if c.isalnum():
            positions.setdefault(c, [])
            positions[c].append((x, y))

distances = {}

for character, pos_list in positions.items():
    for pos in pos_list:
        for j in range(len(pos_list)):
            dis = ((pos[0] - pos_list[j][0]), (pos[1] - pos_list[j][1]))
            distances.setdefault(pos, [])
            if dis != (0, 0):
                distances[pos].append(dis)

marked_puzzle = [list(row) for row in rows]

for pos, dis_list in distances.items():
    for dis in dis_list:
        adjusted_x = pos[0] + dis[0]
        adjusted_y = pos[1] + dis[1]
        if 0 <= adjusted_x < len(rows) and 0 <= adjusted_y < len(rows[adjusted_x]):
            marked_puzzle[adjusted_x][adjusted_y] = "#"

marked_puzzle = ["".join(row) for row in marked_puzzle]

antinode_count = sum(1 for row in marked_puzzle for c in row if c == "#")

print(antinode_count)
```

```python
# Exemplo 2

input = """##....#....#
.#.#....0...
..#.#0....#.
..##...0....
....0....#..
.#...#A....#
...#..#.....
#....#.#....
..#.....A...
....#....A..
.#........#.
...#......##"""

rows = []

for line in input:
    rows.append(line.strip())

positions = {}

for x, row in enumerate(rows):
    for y, c in enumerate(row):
        if c.isalnum():
            positions.setdefault(c, [])
            positions[c].append((x, y))

distances = {}

for character, pos_list in positions.items():
    for pos in pos_list:
        for j in range(len(pos_list)):
            dis = ((pos[0] - pos_list[j][0]), (pos[1] - pos_list[j][1]))
            distances.setdefault(pos, [])
            if dis != (0, 0):
                distances[pos].append(dis)

marked_puzzle_inline = [list(row) for row in rows]

for pos, dis_list in distances.items():
    for dis in dis_list:
        x, y = pos[0], pos[1]
        while 0 <= x < len(rows) and 0 <= y < len(rows[x]):
            marked_puzzle_inline[x][y] = "#"
            x += dis[0]
            y += dis[1]

marked_puzzle_inline = ["".join(row) for row in marked_puzzle_inline]

antinode_count_inline = sum(1 for row in marked_puzzle_inline for c in row if c == "#")

print(antinode_count_inline)
```
